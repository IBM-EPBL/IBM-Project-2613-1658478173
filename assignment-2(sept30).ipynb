```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.figure(figsize=(9,9))
plt.title("churn modelling")
plt.xlabel('CreditScore',fontsize=20)
plt.ylabel('EstimatedSalary',fontsize=20)
sns.stripplot(x=d['CreditScore'],y=d['EstimatedSalary'],color='b')
plt.show()
```

```python
plt.hist(d["Age"])
plt.show()
```

bivariate

```python
plt.figure(figsize=(9,9))
plt.scatter(d.Age, d.Tenure)
plt.show()
```

multivariate

```python
plt.figure(figsize=(11,20))
ax = d[["Age","Tenure","CreditScore"]].plot()
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap( d, linewidth = 1 , annot = True)
plt.title( "HeatMap using Seaborn Method" )
plt.show()
```
